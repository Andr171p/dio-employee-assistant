# Пайплайн для загрузки и обработки PPTX -файлов

```python
from langchain_community.document_loaders import UnstructuredPowerPointLoader
```

Путь до pptx файла:

```python
import os

BASE_DIR: str = r"C:\Users\andre\IdeaProjects\DIORag"

DOCUMENTS_DIR: str = os.path.join(BASE_DIR, "documents")

FILE_PATH: str = os.path.join(BASE_DIR, "knowledge_base", "Презентации", "навыки_деловой_переписки.pptx")
```

```python
loader = UnstructuredPowerPointLoader(FILE_PATH)

documents = loader.load()

documents
```

Запись в txt файл:

```python
FILENAME: str = "навыки_деловой_переписки.txt"

file_path = os.path.join(DOCUMENTS_DIR, FILENAME)

with open(
    file=file_path,
    mode="w",
    encoding="utf-8"
) as file:
    for document in documents:
        file.write(document.page_content + "\n")
```
