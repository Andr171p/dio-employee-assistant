```python
from langchain_community.document_loaders import Docx2txtLoader
```

```python
import os

BASE_DIR: str = r"C:\Users\andre\IdeaProjects\DIORag"

DOCUMENTS_DIR: str = os.path.join(BASE_DIR, "documents")
```

```python
FILE_PATH = r"C:\Users\andre\IdeaProjects\DIORag\knowledge_base\КП\Комерческое_предложение_НГХК_УАТ.docx"
```

```python
loader = Docx2txtLoader(FILE_PATH)

documents = loader.load()

documents
```

```python
FILENAME: str = "Комерческое_предложение_НГХК_УАТ.txt"

file_path = os.path.join(DOCUMENTS_DIR, FILENAME)

with open(
    file=file_path,
    mode="w",
    encoding="utf-8"
) as file:
    for document in documents:
        file.write(document.page_content + "\n")
```
